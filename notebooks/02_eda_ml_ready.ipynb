```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_ml = pd.read_csv("../data/final/EV_ICE_ML_READY.csv")
df_ml.head()
```

```python
df_ml.shape
df_ml.info()
```

```python
numeric_df = df_ml.select_dtypes(include=["int64", "float64"])
plt.figure(figsize=(10,6))
sns.heatmap(
    numeric_df.corr(),
    cmap="coolwarm",
    annot=False
)
plt.title("Feature Correlation Matrix (Numeric Features)")
plt.show()
```
